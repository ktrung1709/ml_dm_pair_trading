```python
import requests
from bs4 import BeautifulSoup
import re

import pandas as pd
```

```python
df = pd.read_excel('ticker.xlsx', sheet_name=0) # can also index sheet by name or fetch all sheets
tickerList = df['Ticker'].tolist()
# print(tickerList)

industryList = []
```

```python
count = 0
for ticker in tickerList:
# Specify the URL of the HTML link
    url = 'https://www.marketwatch.com/investing/stock/'+ ticker +'/financials?countrycode=vn&mod=mw_quote_tab'
    # Send a GET request to the URL and retrieve the HTML content
    response = requests.get(url)
    html_content = response.text

    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(html_content, 'html.parser')


    # Extract the desired data from the HTML
    # You can use various methods provided by BeautifulSoup to navigate and search the HTML structure
    # For example, to extract all the text from <p> tags:
    # h3s = soup.find_all('h3')
    # data = [h3.get_text() for h3 in h3s]

    # # Process or analyze the extracted data as per your requirements
    # # For example, print the extracted text from all <p> tags
    # for text in data:
    #     print(text)
    html_string = soup.prettify()
    # Specify the file path where you want to save the HTML content
    file_path = 'result.html'

    # Open the file in write mode and write the HTML content
    with open(file_path, 'w', encoding='utf-8') as file:
        file.write(html_string)


    # Load the HTML file
    with open(file_path, 'r', encoding='utf-8') as file:
        html_content = file.read()

    # Create a BeautifulSoup object
    soup = BeautifulSoup(html_content, 'html.parser')



    # Find the script tag containing the desired code
    script_tag = soup.find('script', string=lambda string: 'googletag.pubads().setTargeting' in string)

    if script_tag:
        # Extract the industry value using regular expressions
        industry_match = re.search(r"googletag\.pubads\(\)\.setTargeting\('sector', \['(.*?)'\]\);", script_tag.string)
        if industry_match:
            industry = industry_match.group(1)
            print("Industry:", industry)
            industryList.append(industry)
        else:
            print("Industry not found")
            industryList.append('')
        count+=1
        print(count)
    else:
        print("Script tag not found")
        industryList.append('')
        count+=1
        print(count)
```

```python
df['Sector'] = industryList
df.to_excel("sector.xlsx")
print("Done")
```
